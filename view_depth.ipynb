```python
from matplotlib import pyplot as plt
import numpy as np
import pathlib

s2d3d_root = pathlib.Path(r'/Volumes/Elements/sf2d3d_pano')
print(s2d3d_root)
depth_dir = s2d3d_root / 'depth'
depth_png_list = list(depth_dir.glob('*.png'))
print(len(depth_png_list))
```

```python
from PIL import Image
depth = np.array(Image.open(depth_png_list[0]))
mask = depth == 65535
depth = depth.astype(np.float32) / 512.0
depth[mask] = 0
print(depth.min(), depth.max(), depth.mean())
plt.hist(depth.flatten(), bins=100, range=(0,10));plt.show()
```
